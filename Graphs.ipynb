```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fine tune vs Frozen Weights Resnet
finetune_acc = np.load("./resnet_finetune_valid_acc.npy")
finetune_acc_dirt = np.load("./resnet_finetune_valid_dirty_acc.npy")

weighted_acc = np.load("./resnet_valid_acc.npy")
weighted_acc_dirt = np.load("./resnet_valid_dirty_acc.npy")

fig, ax = plt.subplots(1)
ax.set_xlabel("Epoch")
ax.set_ylabel("Accuracy")

ax.plot(range(len(finetune_acc)), finetune_acc, label="Fine Tuned Accuracy", color = '#49d8fc')
ax.plot(range(len(finetune_acc_dirt)), finetune_acc_dirt, label="Fine Tuned Dirty Accuracy", linestyle = 'dotted', color = '#49d8fc',)

ax.plot(range(len(weighted_acc)), weighted_acc, label="Frozen Weights Accuracy", color = '#fcb249')
ax.plot(range(len(weighted_acc_dirt)), weighted_acc_dirt, label="Frozen Weights Dirty Accuracy", linestyle = 'dotted', color = '#fcb249')

ax.set_ylim((0.6,1))
ax.legend()
plt.title("Fine Tuned and Frozen Weight Resnets CE vs. Epochs")
```

```python
# Weighted vs Normal Class Weights

finetune_acc = np.load("./resnet_finetune_valid_acc.npy")
finetune_acc_dirt = np.load("./resnet_finetune_valid_dirty_acc.npy")

weighted_acc = np.load("./resnet_weightedClasses_valid_acc.npy")
weighted_acc_dirt = np.load("./resnet_weightedClasses_valid_dirty_acc.npy")

fig, ax = plt.subplots(1)
ax.set_xlabel("Epoch")
ax.set_ylabel("Accuracy")

ax.plot(range(len(finetune_acc)), finetune_acc, label="Normal Weights Accuracy", color = '#49d8fc')
ax.plot(range(len(finetune_acc_dirt)), finetune_acc_dirt, label="Normal Weights Dirty Accuracy", linestyle = 'dotted', color = '#49d8fc',)

ax.plot(range(len(weighted_acc)), weighted_acc, label="Weighted Accuracy", color = '#fcb249')
ax.plot(range(len(weighted_acc_dirt)), weighted_acc_dirt, label="Weighted Dirty Accuracy", linestyle = 'dotted', color = '#fcb249')

# ax.set_ylim((0.7,1))
ax.legend()
plt.title("Fine Tuned and Frozen Weight Resnets CE vs. Epochs")
```
